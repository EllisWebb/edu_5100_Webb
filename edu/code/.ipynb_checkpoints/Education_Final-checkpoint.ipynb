# Introduction

#### In this notebook, we will analyze how a set of socioeconomic predictive variables can help us better understand patterns in average ACT scores across educational institutions in the United States in 2016-2017. Our goal: to explore which community level factors could be most closely associated with student performance, and thus, to using these predictors to predict ACT outcomes, assesing thier accuracy where applicable.

Prediction variable:

	•	Average ACT Score (or SAT equivalent) | average_act (by school)


Predictive variables:

	•	Percent Adults with College Degree | percent_college (by census tract)
	•	Unemployment Rate | rate_unemployment (by census tract)
	•	Percent Children in Married-Couple Families | percent_married (by census tract)
	•	Median Household Income | median_income (by census tract)
	•	Percent Free and Reduced Lunch | percent_lunch (by school)
    •	Average Teacher Salary per Pupil | salary_pupil (by school district)

Identifying variables:

	•	School Year | year
	•	School ID | id
	•	District ID | l_id
	•	State | state
	•	Zip Code | zip_code
	•	School Type | school_type
	•	School Level | school_level
	•	Charter Status | charter

Data Sources

	1.	EdGap.org: 
Providing census-tract-level socioeconomic data with associated averge school ACT perormance. 
    
	2.	The IES National Center for Education Statistics (NCES): 
Providing district and school identifying data and district level financial information.

It should be noted that EdGap data limits the scope of this analysis to 20 U.S. states.

#### With this data, we will conduct typical cleaning and joining procedures to produce a single, comprehensive dataset. From there, we will impute missing values using an iterative imputer, untilizing the power of our socioeconomic variables and their associated predictors. Finally, we will perform Single and Multiple Linear Regression (MLR) analyses, comparing their accuracy and predictive power using Mean Absolute Error (MAE) & R - squared values. 
    

# Data Prep

## Lets Start by Cleaning our edgap data set

This data set represents data from 20 states and records a series of socioeconomic variables stated above, as well as ACT averages by school. 

### Snake Casing

```python
edgap = edgap.rename(
    columns={
        "NCESSCH School ID": "id",
        "CT Pct Adults with College Degree": "percent_college",
        "CT Unemployment Rate": "rate_unemployment",
        "CT Pct Childre In Married Couple Family": "percent_married",
        "CT Median Household Income": "median_income",
        "School ACT average (or equivalent if SAT score)": "average_act",
        "School Pct Free and Reduced Lunch": "percent_lunch",
    }
)
```

## Now Lets Tackle Cleaning our edu_info data set

This data set represents a series of school identifying variables that we will use later to trim down the data to high schools only. 

### Simplifying and Snake Casing

```python
edu_info = edu_info[
    ['SCHOOL_YEAR', 'NCESSCH', 'LEAID', 'LSTATE', 'LZIP', 'SCH_TYPE_TEXT', 'LEVEL', 'CHARTER_TEXT'
    ]
]
```

```python
edu_info = edu_info.rename(
    columns={
        "SCHOOL_YEAR": "year",
        "NCESSCH": "id",
        "LEAID": "l_id",
        "LSTATE": "state",
        "LZIP": "zip_code",
        "SCH_TYPE_TEXT": "school_type",
        "LEVEL": "school_level",
        "CHARTER_TEXT": "charter"
    }
)
```

### Turning ID's From Float to Objects for Joining

```python
edu_info[['id', 'l_id']] = edu_info[['id', 'l_id']].astype('object')
```

## Next Lets Clean our pupil data set

This data set represents. 

### Simplifying and Snake Casing

```python
pupil = pupil[
    [ "Agency ID - NCES Assigned [District] Latest available year", "Total Current Expenditures - Salary (Z32) per Pupil (V33) [District Finance] 2016-17"
    ]
]
```

```python
pupil = pupil.rename(
    columns={
        "Agency ID - NCES Assigned [District] Latest available year": "l_id",
        "Total Current Expenditures - Salary (Z32) per Pupil (V33) [District Finance] 2016-17": "salary_pupil"
    }
)
```

Our salary per pupil is currently an object, we will need it as a float for future processing and analysis. 

```python
pupil['salary_pupil'] = pd.to_numeric(pupil['salary_pupil'], errors='coerce')
```

On the other hand, we want our l_id as an object to join. 

```python
pupil[['l_id']] = pupil[['l_id']].astype('object')
```

## Begin Merging/joining data set

We want out ACT values to be prioritized and kept in full, data that is not associated with ACT scores, is much less useful for our purpouses and can be dropped. Start with edgap on the "far-left" and join with our edu_info.

```python
ed_merge = edgap.merge(
    edu_info,
    how='left',
    on='id'
)
```

Then sandwich the edu_info in the "middle" by joining our new ed_merge data set with the pupil data set leveraging the l_id value in the edu_info to tie district averaged salary/pupil values to schools within that district. 

```python
edu = ed_merge.merge(
    pupil,
    how='left',
    on='l_id'
)
```

# Sanity Checks

```python
edu.select_dtypes(include=['number']).agg(['min', 'max']).round(2)
```

Let's Remove Unreasonable Numbers. 

For example, negative percentages for lunch, values less the 1 on the act, and 0 dollar spending per pupil can all be switched to NaN to avoid skewing out data. 

```python
edu.loc[edu['percent_lunch'] < 0, 'percent_lunch'] = np.nan
```

```python
edu.loc[edu['average_act'] < 1, 'average_act'] = np.nan
```

```python
edu.loc[edu['salary_pupil'] == 0, 'salary_pupil'] = np.nan
```

Let's Limit our school to High-Schools as Others Dont have Applicable ACTs 

```python
edu = edu.loc[edu['school_level'] == 'High']
```

Identifying Major Gaps in the Data

```python
edu.isna().sum().to_frame(name='Number of Missing Values')
```

For Simplicity we Will Drop our ACT NaNs as There Are Only 3. Some states not reporting salary/pupil data may explain why there are so many NaNs. This may influence imputations. 

```python
edu = edu.dropna(subset=['average_act'])
```

### Lets Impute  

```python
predictor_variables = [
    'rate_unemployment',
    'percent_college',
    'percent_married',
    'median_income',
    'percent_lunch',
    'state',
    'charter',
    'salary_pupil'
]

imputer = IterativeImputer()
```

These predictor variables can now be used as the columns that the imputer will reference for the best guess imputation method.

```python
numerical_predictors = edu[predictor_variables].select_dtypes(include='number').columns.to_list()

print(numerical_predictors)
```

Now we fit the imputer using these columns. Fitting allows the imputer to learn about the relationships between our preidctors and create a more accurate guess. 

```python
imputer.fit(edu.loc[:, numerical_predictors])
```

We now will take the imputer we have trained the imputer off of and use it to impute the NaN values. 

```python
edu.loc[:, numerical_predictors] = imputer.transform(edu.loc[:, numerical_predictors])
```

```python
edu.isna().sum().to_frame(name='Number of Missing Values')
```

```python
edu.to_csv(
    'education_clean.csv',
    encoding='utf-8-sig',
    index=False
)
```

It should be noted that at least 558 of the origional 766 missing salary values came exclusively from NY & WA. Neither of which reported salaries by pupil during the year of 2016-2017. We are assuming that the predictive variables are correlated enough to completely impute those states values based on other information.

# Modeling

We begin with a Full Multiple Linear Regression to feel out what predictors will help the most. 

```python
mlr_full = smf.ols(
    formula='average_act ~ rate_unemployment + percent_college + percent_married + median_income + percent_lunch + salary_pupil',
    data=edu).fit()

print(mlr_full.summary())
```

```python
mean_absolute_error(edu['average_act'], mlr_full.predict())
```

This is quite accurate, meaning we can predict within about 1 act point a student ACT. But we noticed the median income and salary variables may have some curves or nonlinear visuals. Lets test this using an order 2 vs order 1 linear regression. 

```python
plt.figure(figsize=(6, 6))

sns.regplot(data=edu,
            x='median_income',
            y='average_act',
            color='blue',
            ci=False,
            scatter_kws={'color': 'black', 'edgecolors': 'white', 'linewidths': 1})

sns.regplot(data=edu,
            x='median_income',
            y='average_act',
            order=2, ## This is where you specify the quadratic model. 
            color='orange',
            ci=False,
            scatter=False)

plt.xlabel('Median income ($)', fontsize=16)
plt.ylabel('Average ACT score', fontsize=16)

plt.tick_params(labelsize=14)

plt.show()
```

```python
plt.figure(figsize=(6, 6))

sns.regplot(data=edu,
            x='salary_pupil',
            y='average_act',
            color='blue',
            ci=False,
            scatter_kws={'color': 'black', 'edgecolors': 'white', 'linewidths': 1})

sns.regplot(data=edu,
            x='salary_pupil',
            y='average_act',
            order=2, ## This is where you specify the quadratic model. 
            color='orange',
            ci=False,
            scatter=False)

# Add axis labels
plt.xlabel('Salary Per Pupil ($)', fontsize=16)
plt.ylabel('Average ACT score', fontsize=16)

# Increase the fontsize of the tick labels
plt.tick_params(labelsize=14)

plt.show()
```

We dont see any major deviation from the lower order regression, so we can be confident in continuing forward without being able to get much more accurate. Since we saw large p values for median income and marriage we can drop those variables as they are not statistically significant in helping predict. Salary doesnt seem to be very correlated, but did have a low p value so lets keep it for now. 

```python
mlr_reduced = smf.ols(
    formula='average_act ~ rate_unemployment + percent_college + percent_lunch + salary_pupil',
    data=edu).fit()

print(mlr_reduced.summary())
```

Comparing the accuracy of our full and reduced models

```python
mae_full = mean_absolute_error(edu['average_act'], mlr_full.predict())
mae_reduced = mean_absolute_error(edu['average_act'], mlr_reduced.predict())

r2_full = mlr_full.rsquared
r2_reduced = mlr_reduced.rsquared

pd.DataFrame({'Mean Absolute Error': [mae_full, mae_reduced],
              'R-squared': [r2_full, r2_reduced]},
              index=['full model', 'reduced model']).round(4)
```

We didnt see a huge decrease in accuracy so lets normalize to see if any of our variables are contributing more than other. 

```python
norm_var = ['rate_unemployment', 'percent_college', 'percent_lunch', 'salary_pupil']

scaled_columns = [var + '_normalized' for var in norm_var]

scaler = StandardScaler().fit(edu[norm_var])

edu[scaled_columns] = scaler.transform(edu[norm_var])
```

```python
edu[scaled_columns].agg(['mean', 'std']).round(3)
```

By running this reduced then normalized model, we can look at the magnitude of the coefficients and those with the larges magnitude (absolute value) are contributing most. 

```python
mlr_norm = smf.ols(
    formula='average_act ~ rate_unemployment_normalized + percent_college_normalized + percent_lunch_normalized + salary_pupil_normalized',
    data=edu).fit()
```

```python
print(mlr_norm.summary())
```

We see that percent lunch is are strongest contributer by far, so lets run a linear regression using just it as a predictor. 

```python
lr = smf.ols(
    formula='average_act ~ percent_lunch_normalized',
    data=edu).fit()
```

```python
print(lr.summary())
```

```python
mae_full = mean_absolute_error(edu['average_act'], mlr_full.predict())
mae_reduced = mean_absolute_error(edu['average_act'], mlr_reduced.predict())
mae_norm = mean_absolute_error(edu['average_act'], mlr_norm.predict())
mae_single = mean_absolute_error(edu['average_act'], lr.predict())

r2_full = mlr_full.rsquared
r2_reduced = mlr_reduced.rsquared
r2_norm = mlr_full.rsquared
r2_single = lr.rsquared

pd.DataFrame({'Mean Absolute Error': [mae_full, mae_reduced, mae_norm, mae_single],
              'R-squared': [r2_full, r2_reduced, r2_norm, r2_single]},
              index=['full model', 'reduced model', 'normalized model', '% lunch only']).round(4)
```

Comparing all of our models we see the full model is most accurate, but our percent lunch is the most practical because of its simplicity.
